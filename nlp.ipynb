```python
import pandas as pd
```

```python
df=pd.read_csv("/content/mail_data.csv")
df
```

```python
df.isna().sum()
```

```python
txt=df.Message
```

```python
txt=txt.str.replace("[^a-zA-Z0-9]+"," ",regex=True)
txt
```

```python
df["Category"].value_counts()
```

```python
df["Category"]=df["Category"].map({"ham":1,"spam":0})
```

```python
df["Category"].value_counts()
```

```python
from nltk.stem import SnowballStemmer
stemmerr=SnowballStemmer("english")
from nltk.tokenize import word_tokenize
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
txt=txt.apply(lambda x:[stemmerr.stem(token.lower()) for token in word_tokenize(x)]).apply(lambda token:" ".join(token))
txt
```

```python
from nltk.corpus import stopwords
nltk.download("stopwords")
sw=stopwords.words("english")
sw
```

```python
txt=txt.apply(lambda x:[token for token in word_tokenize(x) if token not  in sw]).apply(lambda token:" ".join(token))
txt
```

```python
txt=txt.apply(lambda x:[token for token in word_tokenize(x) if len(token)>2]).apply(lambda token:" ".join(token))
txt
```

```python
#vectorization
from sklearn.feature_extraction.text import TfidfVectorizer
vector=TfidfVectorizer()
X=vector.fit_transform(txt)
print(X)
```

```python
vector.get_feature_names_out()
```

```python
y=df["Category"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(random_state=1)
rf.fit(X_train,y_train)
```

```python
y_new=rf.predict(vector.transform(["jurong point crazi avail onli bugi great world..."]))
y_new
```

```python
y_pred=rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
